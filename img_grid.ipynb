```python
import os, sys, re
import random
import matplotlib.pyplot as plt
import PIL
from mpl_toolkits.axes_grid1 import ImageGrid

%matplotlib inline
```

```python
data_path = os.path.abspath("/home/chris/Dropbox/HOTOSM/ALL_IMG_15cm/IMG")
png_ls = []

for i, j, k in os.walk(data_path):
    for img in k:
        if img.endswith(".png"):
            png_ls.append(i + "/" + img)
```

```python
ls_32 = random.choices(png_ls, k = 32)
PIL_32 = []

for i in ls_32:
    img = PIL.Image.open(i)
    PIL_32.append(img)

fig = plt.figure(figsize=(20., 20.))
grid = ImageGrid(fig, 111,
                 nrows_ncols = (4, 8),
                 axes_pad = 0.1,
                 label_mode = "L")

for ax, im in zip(grid, PIL_32):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)

plt.show()
```
